```python
#import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
#website link
base_url='https://www.imdb.com/search/title/?release_date=2021-01-01,2021-12-31&sort=num_votes,desc&start='
film_name=[]
year_rele=[]
certificate=[]
time_duration=[]
genre=[]
rating=[]
meta_score=[]
votes=[]

```

```python
for i in range(1,101,51):
  time.sleep(2)
  req=requests.get(base_url+str(i)+'&ref_=adv_nxt',headers={'Accept-Language':'en-US,en;q=0.5'})
  soup=BeautifulSoup(req.text,'html')

  # Obtain each movie in a list
  movies=soup.findAll('div',attrs={'class':'lister-item-content'})
  # Apply loop
  for movi in movies:


    # Name of movie
    mov_name=movi.find('h3',attrs={'class':'lister-item-header'}).a.text
    film_name.append(mov_name)


    # Year of release
    year=movi.find('span',attrs={'class':'lister-item-year text-muted unbold'}).text
    year_rele.append(year)

    #Certificate
    certi=movi.find('span',attrs={'class':'certificate'})
    if certi is None:
      certificate.append('No data')
    else:
      certificate.append(certi.text)

    # Duration of film
    duration=movi.find('span',attrs={'class':'runtime'}) 
    if duration is None:
      time_duration.append('No data') 
    else:
      time_duration.append(duration.text)

    #Genre
    gen=movi.find('span',attrs={'class':'genre'}).text.strip()
    genre.append(gen) 

    # Rate
    ra=movi.find('div',attrs={'class':'inline-block ratings-imdb-rating'}).strong.text
    rating.append(ra)

    # Metascore
    meta=movi.find('span',attrs={'class':'metascore'})
    if meta is None:
      meta_score.append('No data')
    else:
      meta_score.append(meta.text)


    # Votes  
    vote=movi.find('span',attrs={'name':'nv'}).text
    votes.append(vote)
```

```python
data=pd.DataFrame({'name':film_name,'year':year_rele,'certifi':certificate,'time':time_duration,'Genre':genre,'Vote':votes,'Metascore':meta_score,'Rating':rating})
data
```

```python
data.to_csv('film_webscraped_data.csv',index=False)
```
